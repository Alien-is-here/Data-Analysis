```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
cancelled_perc*100
```

```python
plt.figure(figsize=(5,5))
plt.pie(cancelled_perc,labels=cancelled_perc.index,autopct='%1.2f%%')
plt.title('Booking Cancellation Percentage')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.bar(['cancelled','Not Cancelled'], df['is_canceled'].value_counts().values, color=['red','green'], width=0.5)
plt.xticks(rotation=0)
plt.title('Reservation Cancellation Percentage')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='hotel',hue='is_canceled',data=df, palette="Blues")
plt.title('Reservation Cancellation Percentage')
plt.show
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(resort_hotel.index, resort_hotel['adr'] , label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'] , label='City Hotel')
plt.title('Average Daily Rate (ADR) Over Time')
plt.legend()
plt.show()
```

```python
df['Month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(12,6))
ax1 = sns.countplot(x='Month',hue='is_canceled',data=df, palette="bright")
plt.title('Reservation Cancellation Percentage')
plt.legend('not_cancelled' , 'cancelled')
plt.show
```

```python
adr_by_month = df[df['is_canceled'] == 1].groupby('Month')[['adr']].sum().reset_index()

plt.figure(figsize=(15, 8))
plt.bar(x=adr_by_month['Month'], height=adr_by_month['adr'])  # Explicitly provide x and height
plt.title('ADR for Canceled Bookings by Month')  # More descriptive title
plt.xlabel('Month')  # Label the x-axis
plt.ylabel('Total ADR')  # Label the y-axis
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]

top_10_country = cancelled_data['country'].value_counts().sort_values(ascending=False)[:10]
plt.figure(figsize=(10,6))
plt.bar(top_10_country.index, top_10_country.values)
plt.title('Top 10 Countries with Most Canceled Bookings')
plt.xlabel('Country')
plt.ylabel('Number of Canceled Bookings')
plt
```

```python
plt.figure(figsize=(10,6))
plt.pie(top_10_country, autopct = "%.2f" , labels = top_10_country.index)
plt.title('Top 10 Countries with Most Canceled Bookings')
plt.xlabel('Country')
plt.ylabel('Number of Canceled Bookings')
plt
```
